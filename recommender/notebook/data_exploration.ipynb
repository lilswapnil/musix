```python
plt.figure(figsize=(10, 6))
sns.heatmap(spotify_data.isnull(), cbar=False, yticklabels=False, cmap="viridis")
plt.title("Missing Values Heatmap of Spotify Tracks Dataset")
plt.show()
```

```python
spotify = spotify_data.drop(['Unnamed: 0', 'time_signature'], axis=1)
```

```python
# Split the dataset into training, validation, and test sets
spotify_train = spotify.sample(frac=0.8, random_state=42)
spotify_temp = spotify.drop(spotify_train.index)
spotify_val = spotify_temp.sample(frac=0.5, random_state=42)
spotify_test = spotify_temp.drop(spotify_val.index)
```
